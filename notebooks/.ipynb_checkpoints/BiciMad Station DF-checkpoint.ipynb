```python
# Import pandas 
import pandas as pd
import requests
```

```python
path1 = '../data/bicimad_stations.csv'
df1 = pd.read_csv(path1, sep= "\t")
df1
```

```python
print(type(df1.loc[:,'geometry.coordinates'][0]))
```

```python
df1[['longitud', 'latitud']] = df1['geometry.coordinates'].str.strip('[]').str.split(', ', expand=True).astype(float)
```

```python
print(type(df1.loc[:,'longitud'][0]))
```

```python
df1['name'] = df1['name'].str.extract(r'\s-\s(.*)')
```

```python
df1.loc[:,'name']
```

```python
bicimad_stations = df1[['name', 'address', 'latitud', 'longitud']]
```

```python
bicimad_stations = bicimad_stations.rename(columns={'name': 'BiciMAD station', 'address': 'Station location',
                                                    'latitud': 'station_lat','longitud': 'station_long'})
bicimad_stations
```

```python
statio_null_cols = bicimad_stations.isnull().sum()
statio_null_cols
```
